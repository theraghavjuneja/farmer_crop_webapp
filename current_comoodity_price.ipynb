```python
state_district_map = {}


for index, row in df.iterrows():
    state = row['State']
    district = row['District']
    
    
    if state in state_district_map:
        state_district_map[state].add(district)

    else:
        state_district_map[state] = {district}


for state, districts in state_district_map.items():
    print(f"State: {state}")
    print("Districts:", ", ".join(districts))
    print()
```

```python
import requests

def get_hourly_weather(api_key, city_name):
    url = f"http://api.openweathermap.org/data/2.5/forecast?q={city_name}&appid={api_key}"
    response = requests.get(url)
    data = response.json()
    return data

# Replace '9fbad4ea130c7759ec312350195588c1' with your actual OpenWeatherMap API key
api_key = '9fbad4ea130c7759ec312350195588c1'
city_name = 'New York'  # Replace with the name of the city you want to get weather data for

hourly_weather = get_hourly_weather(api_key, city_name)
print(hourly_weather)
```
